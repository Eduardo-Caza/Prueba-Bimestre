```python
import pandas as pd
peliculas=pd.read_csv("imdb_movie_data_2023.csv")
palabra = 'Comedy'
palabra2= 'Crime'
palabra3= 'Horror'
palabra4= 'Drama'
palabra5= 'History'
palabra6= 'Thriller'
palabra7= 'Adventure'
palabra8= 'Romance'
palabra9= 'Action'
palabra10='Family' 
comedia = peliculas['Genre'].str.contains(palabra).sum()
crime = peliculas['Genre'].str.contains(palabra2).sum()
horror = peliculas['Genre'].str.contains(palabra3).sum()
drama = peliculas['Genre'].str.contains(palabra4).sum()
historia = peliculas['Genre'].str.contains(palabra5).sum()
thriller = peliculas['Genre'].str.contains(palabra6).sum()
aventura = peliculas['Genre'].str.contains(palabra7).sum()
romance = peliculas['Genre'].str.contains(palabra8).sum()
accion = peliculas['Genre'].str.contains(palabra9).sum()
familia = peliculas['Genre'].str.contains(palabra10).sum()
print(f'La palabra "{palabra}" se repite {comedia} veces.')
print(f'La palabra "{palabra2}" se repite {crime} veces.')
print(f'La palabra "{palabra3}" se repite {horror} veces.')
print(f'La palabra "{palabra4}" se repite {drama} veces.')
print(f'La palabra "{palabra5}" se repite {historia} veces.')
print(f'La palabra "{palabra6}" se repite {thriller} veces.')
print(f'La palabra "{palabra7}" se repite {aventura} veces.')
print(f'La palabra "{palabra8}" se repite {romance} veces.')
print(f'La palabra "{palabra9}" se repite {accion} veces.')
print(f'La palabra "{palabra10}" se repite {familia} veces.')
```

```python
fecha = 2023
a=2022
b=2021
c=2020
d=2019
e=2018
f=2017
g=2016
h=2015
i=2014
j=2013
k=2012
l=2011
m=2010
n=2009
o=2008
p=2007
q=2006
r=2005
s=2004
t=2003
u=2002
v=2001
w=2000
año = (peliculas['Year'] == fecha).sum()
año1 = (peliculas['Year'] == a).sum()
año2 = (peliculas['Year'] == b).sum()
año3 = (peliculas['Year'] == c).sum()
año4 = (peliculas['Year'] == d).sum()
año5 = (peliculas['Year'] == e).sum()
año6 = (peliculas['Year'] == f).sum()
año7 = (peliculas['Year'] == g).sum()
año8 = (peliculas['Year'] == h).sum()
año9 = (peliculas['Year'] == i).sum()
año10 = (peliculas['Year'] == j).sum()
año11 = (peliculas['Year'] == k).sum()
año12 = (peliculas['Year'] == l).sum()
año13 = (peliculas['Year'] == m).sum()
año14 = (peliculas['Year'] == n).sum()
año15 = (peliculas['Year'] == o).sum()
año16 = (peliculas['Year'] == p).sum()
año17 = (peliculas['Year'] == q).sum()
año18 = (peliculas['Year'] == r).sum()
año19 = (peliculas['Year'] == s).sum()
año20 = (peliculas['Year'] == t).sum()
año21 = (peliculas['Year'] == u).sum()
año22 = (peliculas['Year'] == v).sum()
año23 = (peliculas['Year'] == w).sum()
palabras = [año, año1, año2, año3, año4, año5, año6, año7, año8, año9, año10, año11, año12, año13, año14, año15, año16, año17, año18, año19, año20, año21, año22, año23]

palabras
```

```python
generos = ['2023', '2022', '2021', '2020', '2019', '2018', '2017', '2016', '2015']
palabras = [año, año1, año2, año3, año4, año5, año6, año7, año8]
plt.bar(generos, palabras)
plt.title('Años de produccion', fontsize = 17,   fontname = 'Tahoma')
plt.xlabel('Años')
plt.ylabel('N° de pEliculas')
plt.xticks(fontname = 'Verdana') 
plt.yticks(fontname = 'Verdana')  
plt.figure(figsize=(20, 16))
plt.show()
```
